#The oxbuild dataset (download and move to the project folder).

##Download the necessary files

**Images**

```python
import requests

link = "https://thor.robots.ox.ac.uk/datasets/oxford-buildings/oxbuild_images-v1.tgz"

response = requests.get(link)
with open("oxbuild_images-v1.tgz", "wb") as f:
    f.write(response.content)
```

**Groundtruth**

```python
import requests

link = "https://www.robots.ox.ac.uk/~vgg/data/oxbuildings/gt_files_170407.tgz"

response = requests.get(link)
with open("gt_files_170407.tgz", "wb") as f:
    f.write(response.content)
```
